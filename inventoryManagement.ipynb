```python
import json
import datetime
fd = open("record.json","r")
r = fd.read()
fd.close()
record = json.loads(r)

f = open("sales.json","r")
s = f.read()
f.close()
sales = json.loads(s)
```

```python
print('''
-----------------------------------------------------
Welcome to Inventory 
-----------------------------------------------------
0. Quit
1. View Items 
2. Add Items
3. Buy Items
4. Delete Items
5. Edit Items
6. Sales
7. Help-See all commands again
''')

choice=-999
while choice!=0:
    choice = int(input("Ente Your Option: "))
    if choice==1:
        print('_____________________________________________________')
        print("Available Items")
        print('_____________________________________________________')
        for keys,values in record.items():
            print("Product ID:",keys)
            for item in values:
                print(item+ ':', values[item])
            print('-----------------------------------------------------')

    elif choice==2:
        add_prod_id = str(input("Enter product id:"))
        add_name = str(input("Enter Product name:"))
        add_size = str(input("Enter Product size:"))
        add_quantity = int(input("Enter Product quantity:"))
        add_company = str(input("Enter Product company:"))
        add_price = int(input("Enter Product price:"))

        for i in range(0,len(record.keys())):
            if(add_prod_id in record.keys()):
                print("This Product is already Exists in Records.")
                e = input('You Want to add quantity in the product record("Y/N"):')
                if(e=='Y' or e=='y'):
                    print()
                    (record[add_prod_id]['quantity']) = (record[add_prod_id]['quantity']) +(add_quantity)
                    print("Quantity Added Successfully:",record[add_prod_id])
                    break
                else:
                    print("Nothing changed in records.")
                break
            else:
                record[add_prod_id] = {'name': add_name, 'size': add_size, 'quantity': add_quantity, 'company': add_company, 'price': add_price}
            break
        js = json.dumps(record)

        fd = open("record.json",'w')
        fd.write(js)
        fd.close()

    elif choice==3:
        ui_prod=str(input("Enter the Prodeuct Id:"))
        ui_quant=int(input("Enter the product quantity:"))

        if (record[ui_prod]['quantity']) >= ui_quant:
            print("-----------------------------------------------------")
            print("Product:", record[ui_prod]['name'])
            print("Price:", record[ui_prod]['price'])
            print("Billing Amount:", record[ui_prod]['price']*ui_quant)
            print("Thank you for purchasing Items.")
            print("-----------------------------------------------------")

            (record[ui_prod]['quantity']) = (record[ui_prod]['quantity']) -(ui_quant)
            

            dt=datetime.datetime.now()

            sales[(len(sales)+1)] = {'Date&Time': dt.strftime("%b %d %Y %H:%M:%S"), 'Product' : ui_prod,'Name' : record[ui_prod]['name'], 'Quantity' : ui_quant, 'Amount' : record[ui_prod]['price']*ui_quant}

            j = json.dumps(sales)
            f = open("sales.json",'w')
            f.write(j)
            f.close()

            js = json.dumps(record)

            fd = open("record.json",'w')
            fd.write(js)
            fd.close()

        elif(record[ui_prod]['quantity']==0):
            print("This Item Is Finished.")
        elif (record[ui_prod]['quantity']) <= ui_quant:
            print("Available quantity is ", record[ui_prod]['quantity'])
            print()
            ans = input("We have only this amount of quantity left.You want to purchase this('Y/N'):")

            if(ans=='y' or ans=='Y'):
                print("Thank you for purchasing Items.")
                dt=datetime.datetime.now()

                sales[(len(sales)+1)] = {'Date&Time': dt.strftime("%b %d %Y %H:%M:%S"), 'Product' : ui_prod, 'Name' : record[ui_prod]['name'], 'Quantity' : record[ui_prod]['quantity'], 'Amount' : record[ui_prod]['price']*record[ui_prod]['quantity']}
        
                j = json.dumps(sales)
                f = open("sales.json",'w')
                f.write(j)
                f.close()

                js = json.dumps(record)

                fd = open("record.json",'w')
                fd.write(js)
                fd.close()
                (record[ui_prod]['quantity']) = 0
            
            else:
                print("Ok")
        else:
            print("Product is not available.")

    elif choice==4:
        del_item = str(input("Enter Delete Item Id:"))
        del record[del_item]
        print("Item Deleted Successfully")
        js = json.dumps(record)

        fd = open("record.json",'w')
        fd.write(js)
        fd.close()

    elif choice==5:
        prod_id = str(input("Enter product id which you want to Update:"))

        print('''
        What you want to update:
        a. Name 
        b. Size
        c. Quantity
        d. Company
        e. Price
        ''')

        update = input("Ente Your Option: ")
        if update=='a':
            edit_name = str(input("Enter Product name to edit:"))
            record[prod_id]['name']=edit_name
            print("Name is Successfully updated")
        elif update=='b':
            edit_size = str(input("Enter Product size to edit:"))
            record[prod_id]['size']=edit_size
            print("Size is Successfully updated")
        elif update=='c':
            edit_quantity = int(input("Enter Product quantity to edit:"))
            record[prod_id]['quantity']=edit_quantity
            print("Quantity is Successfully updated")
        elif update=='d':
            edit_company = str(input("Enter Product company to edit:"))
            record[prod_id]['company']=edit_company
            print("Company is Successfully updated")
        elif update=='e':
            edit_price = int(input("Enter Product price to edit:"))
            record[prod_id]['price']=edit_price
            print("Price is Successfully updated")

        js = json.dumps(record)

        fd = open("record.json",'w')
        fd.write(js)
        fd.close()
    elif choice==0:
        print("Exit Successfully")
        break
    
    elif choice==6:
        print('-----------------------------------------------------')
        for keys,values in sales.items():
            print("Sale ID:",keys)
            for item in values:
                print(item+ ':', values[item])
            print('-----------------------------------------------------')
    
    elif(choice==7):
        print('''
        -----------------------------------------------------
        Welcome to Inventory 
        -----------------------------------------------------
        0. Quit
        1. View Items 
        2. Add Items
        3. Buy Items
        4. Delete Items
        5. Edit Items
        6. Sales
        7. Help-See all commands again
        ''')
    else:
        print("Please enter correct command this command is not available")
```
