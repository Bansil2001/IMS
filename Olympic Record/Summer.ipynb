# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
df = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("Summer Olympics is held in", len(df['City'].unique()) ,"cities")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
medal_data = pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
medal_data.groupby('Sport').count()['Gold'].sort_values(ascending = False).head().plot.bar()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar()
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot.bar()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
medal_data = pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
medal_data.groupby('Athlete').sum()['Gold'].sort_values(ascending = False).head().plot.bar()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
co = df[df['Country'] == 'IND']
med = co[co['Medal'] == 'Gold']
print("India won its first Gold Medal in", med['Year'].values[0],"in",med['City'].values[0],"in",med['Sport'].values[0],"sport.")
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])


pd.DataFrame(data,columns = ['Event','Player']).sort_values(by='Player', ascending=False).head().plot(x = 'Event', y = 'Player', kind = 'bar', figsize = (5,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
ge = df[df['Gender'] == 'Women']
me = pd.concat([ge,pd.get_dummies(ge['Medal'])],axis=1)
me.groupby('Sport').sum()['Gold'].sort_values(ascending = False).head().plot.bar()
```
